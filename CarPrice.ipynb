```python
import datetime
curr_time = datetime.date.today()
data['Current_Year'] = curr_time.year

# To Calculate how old the car is, I created new feature "No_of_Years"
data['No_of_Years'] = data['Current_Year'] - data['Year']

data.head()
```

```python
final_df = data.copy()
final_df.drop(labels= ['Car_Name','Year', 'Current_Year'], axis= 1, inplace= True)
final_df.head()
```

```python
final_df = pd.get_dummies(final_df, drop_first=True)
final_df.head()
```

```python
# Let's check data types of variables
final_df.dtypes
```

```python
# Converting the datatypes of variables as of required datatype
final_df['Fuel_Type_Diesel'] = final_df['Fuel_Type_Diesel'].astype('int64')
final_df['Fuel_Type_Petrol'] = final_df['Fuel_Type_Petrol'].astype('int64')
final_df['Seller_Type_Individual'] = final_df['Seller_Type_Individual'].astype('int64')
final_df['Transmission_Manual'] = final_df['Transmission_Manual'].astype('int64')
```

```python
final_df.head(10)
```

```python
X = final_df.iloc[:, 1:]
y = final_df.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# To check important feature
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Mean Squareed Error: ', mean_squared_error(y_test, y_pred))
print('Root Mean Square Error: ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('Accuracy : ',r2_score(y_test, y_pred)*100)
```
